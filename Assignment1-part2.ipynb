```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def blend_images_manually(img1_path, img2_path, alpha):
    # Validate alpha
    if not (0 <= alpha <= 1):
        raise ValueError("Alpha must be between 0 and 1.")

    # Load both images
    img1 = cv2.imread(img1_path).astype(np.float32)
    img2 = cv2.imread(img2_path).astype(np.float32)

    # Resize img1 to match img2 if needed
    if img1.shape != img2.shape:
        img1 = cv2.resize(img1, (img2.shape[1], img2.shape[0]))

    # Manual blending
    blended = (1 - alpha) * img1 + alpha * img2
    blended = np.clip(blended, 0, 255).astype(np.uint8)

    # Save and display
    cv2.imwrite("manual_blend.jpg", blended)
    plt.imshow(cv2.cvtColor(blended, cv2.COLOR_BGR2RGB))
    plt.title(f'Manual Blend (alpha={alpha})')
    plt.axis('off')
    plt.show()

blend_images_manually('img1.jpg', 'img2.jpg', 0.5)
```
